```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots
import seaborn as sns
import textwrap
from sklearn.preprocessing import StandardScaler
import glob
import numpy as np
import statsmodels.formula.api as smf
import glob
from scipy import stats
```

```python
from presidency_comparison_functions import visual, standardizer, classify_sector, prepare_data, analyze_sector_groups, proportions
```

```python
# Importing total tariffs data (2017-2024) at individual HTS8 level
data = pd.read_excel("DataWeb-Query-Export (2).xlsx", skiprows=2, dtype={'HTS Number': str})
data = data[data.columns[1:]]

###############################################################################################################

# HTS8 -> NAICS
ct = pd.read_excel('impconcord25.xlsx', usecols=['commodity', 'naics'], dtype={'commodity': str})
# Slicing codes to get hts8 and naics.
ct.commodity = ct.commodity.str[:8]
ct.naics = ct.naics.str[:4]

###############################################################################################################

# Creating a hash table whose keys are HTS8 codes and values are NAICS codes
conversion_dict = {}
for _, row in ct.iterrows():
    conversion_dict.setdefault(row['commodity'], row['naics'])

###############################################################################################################

naics_desc = pd.read_excel("naics_description.xlsx", dtype={'2022 NAICS US   Code': str}).set_index('Seq. No.')
naics_desc = naics_desc.drop('Description', axis=1)
description_dict = {}
for _, row in naics_desc.iterrows():
    naics = row['2022 NAICS US   Code']
    description = row['2022 NAICS US Title']
    if len(naics) == 3:
        if description.endswith('T'):
            description = description.rstrip('T')
        description_dict.setdefault(naics,description)
    else:
        continue
```

```python
# Creating actual NAICS and SECTOR columns and dropping useless ones
data['naics4'] = data['HTS Number'].map(conversion_dict) # Useful in the final regression only
data['NAICS'] = data['naics4'].astype(str).apply(lambda x: x[:3])
data['NAICS2'] = data['naics4'].astype(str).apply(lambda x: x[:2])
data['SECTOR'] = data['NAICS'].map(description_dict)
```

```python
# Removing exceptional tariffs since we do not have sufficient information to classify them
exceptions = ['98', '99', 'na']
filt = (data['NAICS2'].isin(exceptions))
data = data.loc[~filt]
```

### REGRESSION MODEL WITH NAICS4

```python
df_analysis3 = prepare_data(df, groups='naics3')
formula3 = 'resid_two_way_z ~ C(Presidency) + C(Presidency):C(is_tech_hardware) + C(Presidency):C(is_traditional)'
# + 'GDP' + 'NumFirms' + 'employment' + 'exposure to trade' + 'lobbying'
model3 = smf.ols(formula=formula3, data=df_analysis3).fit()
results3 = model3.summary()
results3
```

#### VISUALIZATION of REGRESSION RESULTS when NAICS3 are used

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 5)

# Effetti netti per settore e presidenza
# Calcolati da: Intercept + Trump_baseline + Interazioni
data = pd.DataFrame({
    'Sector': ['Tech Hardware', 'Tech Hardware', 'Traditional Mfg', 'Traditional Mfg',
               'Other Sectors', 'Other Sectors'],
    'Presidency': ['Trump', 'Biden', 'Trump', 'Biden', 'Trump', 'Biden'],
    'Effect': [-0.258, 0.582, -0.715, 1.039, 0.325, -0.325]
})

# BOXPLOT

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot per presidenza e settore
sector_order = ['Tech Hardware', 'Traditional Mfg', 'Other Sectors']
palette = {'Trump': 'coral', 'Biden': 'steelblue'}

sns.barplot(data=data, x='Sector', y='Effect', hue='Presidency', 
            palette=palette, ax=ax1, order=sector_order)

ax1.axhline(y=0, color='black', linestyle='-', linewidth=1.5)
ax1.set_xlabel('Sector', fontsize=12, fontweight='bold')
ax1.set_ylabel('Effect on Tariffs (std deviations)', fontsize=12, fontweight='bold')
ax1.set_title('Tariff Effects by Sector and Presidency', fontsize=14, fontweight='bold')
ax1.legend(title='Presidency', fontsize=11)
ax1.grid(axis='y', alpha=0.3)

# Aggiungi valori sulle barre
for container in ax1.containers:
    ax1.bar_label(container, fmt='%.2f', fontsize=10, fontweight='bold')

# DIFFERENCE-IN-DIFFERENCES

# Prepara dati per line plot
tech_data = data[data['Sector'] == 'Tech Hardware'].sort_values('Presidency')
trad_data = data[data['Sector'] == 'Traditional Mfg'].sort_values('Presidency')

x = [0, 1]  # Trump=0, Biden=1
x_labels = ['Trump\n(2017-2020)', 'Biden\n(2021-2024)']

ax2.plot(x, [tech_data.iloc[1]['Effect'], tech_data.iloc[0]['Effect']], 
         'o-', linewidth=3, markersize=12, label='Tech Hardware', color='#2E86AB')
ax2.plot(x, [trad_data.iloc[1]['Effect'], trad_data.iloc[0]['Effect']], 
         's-', linewidth=3, markersize=12, label='Traditional Mfg', color='#A23B72')

ax2.axhline(y=0, color='black', linestyle='--', linewidth=1, alpha=0.5)
ax2.set_xticks(x)
ax2.set_xticklabels(x_labels, fontsize=11)
ax2.set_ylabel('Effect on Tariffs (std deviations)', fontsize=12, fontweight='bold')
ax2.set_title('Parallel Trends: No Preferential Treatment', fontsize=14, fontweight='bold')
ax2.legend(fontsize=11, loc='best')
ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()

# DIFFERENCES BIDEN - TRUMP

fig, ax = plt.subplots(figsize=(10, 6))

# Calcola differenze
sectors = ['Tech Hardware', 'Traditional Mfg', 'Other Sectors']
differences = []
for sector in sectors:
    biden_val = data[(data['Sector'] == sector) & (data['Presidency'] == 'Biden')]['Effect'].values[0]
    trump_val = data[(data['Sector'] == sector) & (data['Presidency'] == 'Trump')]['Effect'].values[0]
    differences.append(biden_val - trump_val)

colors = ['steelblue' if d > 0 else 'coral' for d in differences]

bars = ax.barh(sectors, differences, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
ax.axvline(x=0, color='black', linestyle='-', linewidth=2)
ax.set_xlabel('Biden - Trump (std deviations)', fontsize=12, fontweight='bold')
ax.set_title('How Much Biden Increased Tariffs vs Trump', fontsize=14, fontweight='bold')
ax.grid(axis='x', alpha=0.3)

# Aggiungi valori
for i, (bar, val) in enumerate(zip(bars, differences)):
    x_pos = val + 0.05 if val > 0 else val - 0.05
    ax.text(x_pos, i, f'+{val:.2f}' if val > 0 else f'{val:.2f}', 
            va='center', ha='left' if val > 0 else 'right', 
            fontweight='bold', fontsize=11)

plt.tight_layout()
plt.show();
```

### STACKPLOT

```python
df_analysis_groups = prepare_data(df, groups='sector')
```

```python
# STACKPLOT -> see tariff composition changes over the years

df_analysis_prop = proportions(df_analysis_groups, group_col='SECTOR_GROUP', year_col='Year', value_col='Tariff')
df_pivot = df_analysis_prop.pivot(index='Year', columns='SECTOR_GROUP', values='Tariff_Proportion')
df_pivot = df_pivot.sort_index()

labels = ['Agriculture & Primary', 'High Protection Sectors', 'Mining & Construction',
          'Other Manufacturing','Tech Hardware','Traditional Manufacturing']
data_per_group = {}
for group in labels:
    filt = (df_analysis_prop['SECTOR_GROUP'] == group)
    filtered_df = df_analysis_prop.loc[filt]
    list_tot = []
    for _, row in filtered_df.iterrows():
        tariff = row['Tariff_Proportion']
        list_tot.append(tariff)
    if len(list_tot) != 8:
        print(f'Error occurred with {group}: the list length is {len(list_tot)}.')
    data_per_group.setdefault(group, list_tot)

plt.figure(figsize=(12, 6))
years = ['2017','2018','2019','2020','2021','2022','2023','2024']
plt.stackplot(years,
              data_per_group[labels[0]],
              data_per_group[labels[1]],
              data_per_group[labels[2]],
              data_per_group[labels[3]],
              data_per_group[labels[4]],
              data_per_group[labels[5]],
              labels=labels,
              alpha=0.8)

plt.xlabel('Year', fontweight='bold')
plt.ylabel('Proportion of Total Tariffs', fontweight='bold')
plt.title('Tariff Composition Over Time (Proportional)', fontweight='bold')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show();
```
