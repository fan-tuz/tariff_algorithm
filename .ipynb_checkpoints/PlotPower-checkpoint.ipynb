```python
import numpy as np
from matplotlib.pyplot import subplots
```

```python
def PlotPower(x, p):
    fig, ax = subplots(figsize=(8,8))
    y = x**p
    ax.set_xlabel('Some integers', fontsize=18)
    ax.set_ylabel(f'x^{p}', fontsize=18)
    ax.set_title('A simple function', fontsize=20)

    ax.set_xscale('log') # Setting logarithmic scale

    ax.plot(x, y) ;
```

```python
PlotPower(np.arange(1, 11), 5) # x ranges from 1 to 10, 11 is not included!
```
