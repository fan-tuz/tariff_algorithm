Simple rate of return & log rr (ADDITIVE meaning I can sum log rr for consecutive periods thus avoiding rounding issues, normally distributed)

Daily log rate of return * 365 = yearly log rr

(simple daily rr) ** 365 - 1 = simple yearly rr

```python
# SLOW

sr = np.empty(len(PG))
for i in range(1,len(PG)):
    sr[i]= PG.iat[i,0]/PG.iat[i-1,0] - 1 # As we do not have row -1, I have to start from i=1
```

```python
### vectorization
PG['simple daily return'] = PG['Close']/PG['Close'].shift(1)-1
```

```python
PG['log daily return'] = np.log(PG['Close'])-np.log(PG['Close'].shift(1))
PG['log daily return'].mean()
```

```python
PG['log daily return'].plot(figsize=(15,8))
```
