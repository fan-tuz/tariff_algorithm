```python
import numpy as np
import pandas as pd
from matplotlib.pyplot import subplots

import statsmodels.api as sm

from statsmodels.stats.outliers_influence \
import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
summarize,
poly)
```

```python
df = pd.read_csv('Auto.csv', na_values=['?'])
```

```python
df.dropna(inplace=True)
```

```python
for col in df.columns[1:]:
    template = f"Column {col} has {np.isnan(df[col]).mean()*100}% missing values"
    print(template)
```

```python
y = df['mpg']
X = MS(df.columns[1:]).fit_transform(df)   
```

```python
model = sm.OLS(y,X)
results = model.fit()
summarize(results)
```

##### Comment: even tough cylinders has a low significance, it is included in the model because it is significant when interacting with year, as I will explore later on in the model

```python
predictors = df.columns.drop('mpg') # Remember to remove response variable
for col in predictors:
    remove1 = predictors.drop(col) 
    
    X1 = MS(remove1).fit_transform(df)
    model1 = sm.OLS(y, X1)
    results1 = model1.fit()
    
    print('\n', anova_lm(results1, results),'\n')

    print(f'\nComparison between model statistics for {col} (leaving 1 predictor out first):\n\nR^2: ', results1.rsquared, 'vs', results.rsquared,
          '\nRSE: ', np.sqrt(results1.scale), 'vs', np.sqrt(results.scale),
          '\n\n', '-'*80)
```

#### Impactful predictors: weight, year, origin -> both stats worsen without these predictors
#### Negligible impact: horsepower, displacement, cylinders

```python
results.params['year'] # High value of coefficient and high t-stat suggesting a low standard error
```

```python
vals = [VIF(X, i)
        for i in range(1, X.shape[1])]

vif = pd.DataFrame({'VIF':vals},
                   index=X.columns[1:])
vif

# Let's drop 'displacement' at the start of the code due to its VIF = 22 and "horsepower" because redundant
```

```python
ax = subplots(figsize=(8,8))[1]
ax.scatter(results.fittedvalues, results.resid)
ax.set_xlabel('Fitted value')
ax.set_ylabel('Residual')
ax.axhline(0, c='k', ls='--');

# Evidence of non-linearity
```

```python
infl = results.get_influence()
ax = subplots(figsize=(8,8))[1]

ax.scatter(np.arange(X.shape[0]), infl.hat_matrix_diag) # Leverage statistic for each observation.

ax.set_xlabel('Index')
ax.set_ylabel('Leverage')
np.argmax(infl.hat_matrix_diag)
```

#### One-hot encoding of the categorical variable

```python
# 'origin' and 'cylinders' can become dummy variables. 1st, convert into string dtype
df['origin'] = df['origin'].astype(str)
df = pd.get_dummies(df, columns=['origin'], drop_first=True, dtype=int) #  turns STRING categories into numeric dummy variables as X must be fully numeric to be used in OSL
```

```python
df.dtypes # Essential to check whether dummy cols are numeric
```

```python
vrbs = list(df.columns.drop('mpg'))
y = df['mpg']
X = MS(vrbs).fit_transform(df)
model = sm.OLS(y, X)
summarize(model.fit())
```
